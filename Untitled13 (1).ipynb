# Plotting the amount of number of american conventional arts by the year of their making

 First I reduce the df to only the american works of conventional art. Then I determine the avg year of production for every piece and group by the outcome year


```python
filt = df['Culture']=='American'
american_df = df[filt]
american_df['Object Name'].value_counts().head(10)
```

I pick from this list the common conventional arts to filter from the dataframe

```python

american_canvas = american_df[(american_df['Object Name']=='Drawing') |
                              (american_df['Object Name']=='Painting') |
                              (american_df['Object Name']=='Watercolor')]
american_canvas['year_aprox'] = np.floor((american_canvas['Object Begin Date']+american_canvas['Object End Date'])/2) #in order to get one value for the year I take an avarege of the start and end date
american_canvas_gr = american_canvas.groupby('year_aprox').agg(list).reset_index() #grouping by each available year

```

Counting how many works were made per year

```python
 american_canvas_gr['Object ID'] = american_canvas_gr['Object ID'].apply(lambda x: len(x))
```

Dropping the '0' years

```python
filt = american_canvas_gr['year_aprox']!=0
american_canvas_gr = american_canvas_gr[filt]
```

Plotting the data I seeked

```python
plt.figure(figsize=(20,10))
plt.bar(american_canvas_gr['year_aprox'],american_canvas_gr['Object ID'])
```

# Comparing The amount of Objects from 3 main countries 

```python
main_countries=['England', 'France', 'United States']
df['year_aprox'] = np.floor((df['Object Begin Date']+df['Object End Date'])/2) #enriching the whole dataframe
 
plt.figure(figsize=(10,5))

for i, country in enumerate(main_countries):
  df_c = df[df['Country']==country]
  df_c = df_c.groupby('year_aprox').agg({'Object ID': lambda x: len(list(x))})
  plt.xlim((1000, 2000))
  
  plt.title(country)
  plt.plot(df_c.index,df_c['Object ID'],i)
  
  plt.show()
  plt.figure(figsize=(10,5))
```

# Plotting the works of a popular artist throughout their lifespan

First I will select the most common artist in the museum

```python
artist = df['Artist Display Name'].value_counts().index[0]
artist
```

```python
#filtering the df to only contain Evanses entries
evans_df = df[df['Artist Display Name']==artist]
birth = int(evans_df['Artist Begin Date'].values[0])
#grouping the works by year
evans_grouped = evans_df.groupby('year_aprox').agg({'Object ID':list}).reset_index()
#Adding the age and the number of works she did each year
evans_grouped['n_works'] = evans_grouped['Object ID'].apply(lambda x: len(x))
evans_grouped['age'] = evans_grouped['year_aprox']-birth
# Dropping 0 value
evans_grouped = evans_grouped.drop(0)
```

```python

plt.figure(figsize=(20,10))
plt.title("Timeline of Evans' work")
plt.plot(evans_grouped['age'],evans_grouped['n_works'])
```

# Looking at popular artists most productive age
Some of the works were completed or repaired much after the artists life and their works cant be determined by this df

```python
#filtering the df only to the top artists
a = df[df['Artist Role']=='Artist']
a = a.groupby('Artist Display Name').agg({'Object ID':list,'year_aprox':list,'Artist Begin Date':'first'})
a['n_works'] = a['Object ID'].apply(lambda x: len(x))
a = a.sort_values(by='n_works', ascending=False)
a.head(10)
```

```python
#deciding which artists to present and plotting their most productive year
artists = ['Walker Evans', 'Reginald Marsh', 'Giovanni Battista Piranesi']
a = a[a.index.isin(artists)]
a['Artist Begin Date'] = a['Artist Begin Date'].astype('int64')

a['top year'] = a['year_aprox'].apply(lambda x: int(max(set(x))))
plt.figure(figsize=(20,10))
plt.title('Most productive age of popular artists')
plt.bar(a.index,a['top year']-a['Artist Begin Date'])
```
